## D) Retrieve Data from MySQL DataBase

```python
# import pandas as pd
import mysql.connector as mysql
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

################################################ Define Functions ########################################################
def MySQL_DataBase(table_name):
    sql = "SELECT * FROM {}".format(table_name)
    data = pd.read_sql(sql, con = cnx)
    df = pd.DataFrame(data)
    ### extract data from year 2000 to 2014
    df = df.iloc[10:]
    return df

def data_info(data, db):
    print()
    print("Data Info for {} category: ".format(db))
    print("===============================================================================================")
    print(data.info())
    print()

def table_of_data(data, db):
    print()
    print("Table of Data for {} category: ".format(db))
    print("===============================================================================================")
    print(data)
    print()

def descriptive_stat_summary(data, db, column_name):
    print("Basic Descriptive Statistics Summary for {} category:".format(db))
    print("===============================================================================================")
    print(data[column_name].describe())
    print()
    
    print("Correlation Coefficient:")
    ihd_corr = data.corr(method = "pearson")
    print(ihd_corr)
    print()

def scatter_plot(data, column_name, title, color):
    sns.set(style = "darkgrid", rc = {'figure.figsize':(12,6)})
    ax = sns.regplot(x = 'year', y = column_name, data = data, fit_reg = True, marker = 'D', color = color, label = title)
    ax.set_title("Age-Standardised Mortality Rate for " + title, fontsize = 15, fontweight = "bold")
    ax.set_xlabel("Year", fontweight = "bold", fontsize = 13)
    ax.set_ylabel("Death Per 100,000 Population", fontweight = "bold", fontsize = 13)
    ax.legend(loc = "upper right", facecolor = "white", edgecolor = "navy", shadow = True)

    plt.show(ax)
################################################ Data Analysis ###########################################################

print("***** Welcome to MOH Mortality Rate of Cancer, Stroke and Heart Disease DataBase *****")
print("===============================================================================================")
print("< 1 > Heart Disease")
print("< 2 > Stroke")
print("< 3 > Cancer")
print("-----------------------------------------------------------------------------------------------")
choice = int(input("Enter your choice to view the mortality rate database: "))
print()

user, pw, host = 'root', '83219449dD$$', '127.0.0.1'

if choice == 1:
    db = 'Mortality_rate_IHD'
    table_name = 'heart_diseases'
    column_name = 'ihd'
    title = 'Mortality Rate of Heart Disease'
    color = 'navy'
    
    cnx = mysql.connect(user=user, password=pw, host=host, database=db)
    
    ### call function to extract the required data from MySQL database
    ihd_df = MySQL_DataBase(table_name)
    
    data= ihd_df.copy()
    
    ### call function to display data info
    data_info(data, db)
    
    ### call function to display table of data
    table_of_data(data, db)
    
    ### call function to display basic descriptive stats summary
    descriptive_stat_summary(data, db, column_name)
    
    ### call function to display scatter plot
    scatter_plot(data, column_name, title, color)
    
elif choice == 2:
    db = 'Mortality_rate_Stroke'
    table_name = 'strokes'
    column_name = 'stroke'
    title = 'Mortality Rate of Stroke'
    color = 'tomato'
    
    cnx = mysql.connect(user=user, password=pw, host=host, database=db)
    
    
    ### call function to extract the required data from MySQL database
    ihd_df = MySQL_DataBase(table_name)
    
    data= ihd_df.copy()
    
    ### call function to display data info
    data_info(data, db)
    
    ### call function to display table of data
    table_of_data(data, db)
    
    ### call function to display basic descriptive stats summary
    descriptive_stat_summary(data, db, column_name)
    
    ### call function to display scatter plot
    scatter_plot(data, column_name, title, color)

elif choice == 3:
    db = 'Mortality_rate_Cancer'
    table_name = 'Cancers'
    column_name = 'cancer'
    title = 'Mortality Rate of Cancer'
    color = 'limegreen'
    
    cnx = mysql.connect(user=user, password=pw, host=host, database=db)
    
    ### call function to extract the required data from MySQL database
    ihd_df = MySQL_DataBase(table_name)
    
    data= ihd_df.copy()
    
    ### call function to display data info
    data_info(data, db)
    
    ### call function to display table of data
    table_of_data(data, db)
    
    ### call function to display basic descriptive stats summary
    descriptive_stat_summary(data, db, column_name)
    
    ### call function to display scatter plot
    scatter_plot(data, column_name, title, color)

cnx.close()
```

## Part Two (with no MySQL Included!)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def grouped_bar_chart(mortality_rate_df2):
    sns.set(style = "darkgrid", rc = {'figure.figsize' : (12,6)})
    ax = sns.barplot(x = 'year', y = 'mortality rate', hue = 'variable', data = mortality_rate_df2)
    ax.set_title("Overview of Mortality Rate by Deaths per 100,000 Population", fontsize = 15, fontweight = 'bold')
    ax.set_xlabel("Year", fontsize = 13, fontweight = 'bold')
    ax.set_ylabel("Mortality Rate (Death Per 100,000 Population)", fontsize = 13, fontweight = 'bold')
    ax.legend(facecolor = 'white', edgecolor = 'orchid', shadow = True)
    sns.despine(top  =True, right = True, left = False, bottom = False)
    plt.show(ax)


filename_1 = 'C:\\Users\Jeffrey Wong\SP_Assignment_2\Age-standardised-mortality-rate-for-ischaemic-heart-disease.csv'
filename_2 = 'C:\\Users\Jeffrey Wong\SP_Assignment_2\Age-standardised-mortality-rate-for-stroke.csv'
filename_3 = 'C:\\Users\Jeffrey Wong\SP_Assignment_2\Age-standardised-mortality-rate-for-cancer.csv'

data_1 = pd.read_csv(filename_1, delimiter = ',', na_values = ['','-', 'na', 'NA'])
data_2 = pd.read_csv(filename_2, delimiter = ',', na_values = ['','-','na', 'NA'])
data_3 = pd.read_csv(filename_3, delimiter = ',', na_values = ['', '-', 'na', 'NA'])

df = pd.merge(data_1, data_2, how = 'left', on = 'year')
df = pd.merge(df, data_3, how = 'left', on = 'year')

### extract the data from year 2000 to 2014
mortality_rate_df = df.iloc[10:]
mortality_rate_df2 = pd.melt(mortality_rate_df, id_vars=['year'], value_vars = ['cancer', 'ihd', 'stroke'],
                            var_name = "variable", value_name = "mortality rate")
print(mortality_rate_df2)

### call function to display grouped bar chart
grouped_bar_chart(mortality_rate_df2)
```
