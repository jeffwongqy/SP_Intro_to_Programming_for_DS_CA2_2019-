## Retrieve of Records Directly from CSV Files (No SQL included!)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

################################################# Define Functions #######################################################
def bar_chart(title, copy_data, x_title, y_title, color):
    sns.set(style = "darkgrid", rc = {'figure.figsize' : (12,6)})
    ax = sns.barplot(x = x_title, y = y_title, data = copy_data, color = color, label = "{}".format(y_title))
    ax.set_title("{}".format(title), fontsize = 15, fontweight = 'bold')
    ax.set_xlabel("{}".format(x_title), fontsize = 13, fontweight = 'bold')
    ax.set_ylabel("{}".format(y_title), fontsize = 13, fontweight = 'bold')
    ax.legend(facecolor = 'white', edgecolor = 'navy', shadow = True)
    
    plt.show(ax)

    
def line_chart(title, copy_data, x_title, y_title, color, marker, markerfacecolor):
    sns.set(style = "darkgrid", rc= {'figure.figsize': (12,6)})
    ax = sns.lineplot(x=x_title, y=y_title, data=copy_data, color = color, marker = marker, linewidth = 3, 
                      label = "{}".format(y_title), markerfacecolor = markerfacecolor)
    ax.set_title("{}".format(title), fontsize = 15, fontweight = 'bold')
    ax.set_xlabel("{}".format(x_title), fontsize = 13, fontweight = 'bold')
    ax.set_ylabel("{}".format(y_title), fontsize = 13, fontweight = 'bold')
    ax.legend(facecolor = 'white', edgecolor = 'chocolate', shadow = True)
    
    plt.show(ax)

def merge_data_charts(title, copy_merge_data):
    sns.set(style = "darkgrid", rc= {'figure.figsize': (12,6)})
    ### bar chart for number of claims
    ax1 = plt.subplot2grid((2,2), (0, 0))
    ax1 = sns.barplot(x = copy_merge_data['yr'], y = copy_merge_data['no_of_claims'], label = 'no_of_claims')
    ax1.set_title("Number of Claims", fontsize = 15, fontweight = 'bold')
    ax1.legend(facecolor = 'white', edgecolor = 'navy', shadow = True)
    
    ### bar chart for amount of claims
    ax2 = plt.subplot2grid((2,2), (0, 1))
    ax2 = sns.barplot(x = copy_merge_data['yr'], y = copy_merge_data['amt_claim'], label = 'amt_claim')
    ax2.set_title("Amount of Claims", fontsize = 15, fontweight = 'bold')
    ax2.legend(facecolor = 'white', edgecolor = 'navy', shadow = True)
    
    ### line graph for number of claims
    ax3 = plt.subplot2grid((2,2), (1,0))
    ax3 = sns.lineplot(x = copy_merge_data['yr'], y = copy_merge_data['no_of_claims'], label = 'no_of_claims')
    ax3.set_title("Number of Claims", fontsize = 15, fontweight = 'bold')
    ax3.legend(facecolor = 'white', edgecolor = 'navy', shadow = True)
    
    ### line graph for amount of claims
    ax4 = plt.subplot2grid((2,2), (1,1))
    ax4 = sns.lineplot(x = copy_merge_data['yr'], y = copy_merge_data['amt_claim'], label = 'amt_claims')
    ax4.set_title("Amount of Claims", fontsize = 15, fontweight = 'bold')
    ax4.legend(facecolor = 'white', edgecolor = 'navy', shadow = True)
    
    plt.suptitle("Number of Claims and Amount of Claims for Medishield, Annually")
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    plt.show()
    
#################################################### Data Analysis #######################################################

filename_1 = 'C:\\Users\Jeffrey Wong\SP_Assignment_2\MOH_number_of_claims_medishield.csv'
filename_2 = 'C:\\Users\Jeffrey Wong\SP_Assignment_2\Amount_of_claims_medishield.csv'

no_of_claims = pd.read_csv(filename_1)
amt_of_claims = pd.read_csv(filename_2)

### set both data into dataframe object
no_of_claims_df = pd.DataFrame(no_of_claims)
amt_of_claims_df = pd.DataFrame(amt_of_claims)

print("***** Welcome to MOH Medishield DataBase *****")
print("==================================================================================")
print("< 1 > Number of Claims of Medishield")
print("< 2 > Amount of Claims of Medishield")
print("< 3 > Both Number of Claims of Medishield and Amount of Claims of Medishield")
print("----------------------------------------------------------------------------------")
choice = int(input("Enter your choice <e.g. 2>: "))

if choice == 1:
    title = 'Number of Claims of Medishield'
    
    print()
    print("Data Info for " + title)
    print("-----------------------------------------------------------------------------------------")
    print(no_of_claims_df.info())
    print()
    
    print()
    print("Data for " + title)
    print("-------------------------------------------------------------------------------")
    print(no_of_claims_df)
    print()
    
    print("Basic Descriptive Statistics Summary for " + title)
    print("-------------------------------------------------------------------------------")
    print(no_of_claims_df['no_of_claims'].describe())
    print()
    
    
    print("Graphical Summary for " + title)
    print("-------------------------------------------------------------------------------")
    copy_data = no_of_claims_df.copy()
    x_title = 'yr'
    y_title = 'no_of_claims'
    color = "deepskyblue"
    marker = "D"
    markerfacecolor = "tomato"
    
    ### call the function to display bar chart
    bar_chart(title, copy_data, x_title, y_title, color)
    
    ### call the function to display line chart
    line_chart(title, copy_data, x_title, y_title, color, marker, markerfacecolor)
    
elif choice == 2:
    title = 'Amount of Claims of Medishield'
    
    print()
    print("Data Info for " + title)
    print("-----------------------------------------------------------------------------------------")
    print(amt_of_claims_df.info())
    print()
    
    print()
    print("Data for " + title)
    print("-------------------------------------------------------------------------------")
    print(amt_of_claims_df)
    print()
    
    print("Basic Descriptive Statistics Summary for " + title)
    print("-------------------------------------------------------------------------------")
    print(amt_of_claims_df['amt_claim'].describe())
    print()
    
    
    print("Graphical Summary for " + title)
    print("-------------------------------------------------------------------------------")
    
    copy_data = amt_of_claims_df.copy()
    x_title = 'yr'
    y_title = 'amt_claim'
    color = 'darkorange'
    marker = "o"
    markerfacecolor = "navy"
    
    ### call the function to display bar chart
    bar_chart(title, copy_data, x_title, y_title, color)
    
    ### call the function to display line chart
    line_chart(title, copy_data, x_title, y_title, color, marker, markerfacecolor)
    
elif choice == 3:
    title = 'Number of Claims of Medishield and Amount of Claims of Medishield'
    merge_data = pd.merge(no_of_claims_df, amt_of_claims_df, how = 'right', on = 'yr')
    
    print()
    print("Data Info for " + title)
    print("-----------------------------------------------------------------------------------------")
    print(merge_data.info())
    print()
    
    print()
    print("Data for " + title)
    print("-----------------------------------------------------------------------------------------------------------")
    print(merge_data)
    print()
    
    print("Basic Descriptive Statistics Summary for " + title)
    print("-----------------------------------------------------------------------------------------------------------")
    print(merge_data.iloc[:,1:3].describe())
    print()
    
    print("Graphical Summary for " + title)
    print("-----------------------------------------------------------------------------------------------------------")
    
    copy_merge_data = merge_data.copy()
    
    merge_data_charts(title, copy_merge_data)
```
